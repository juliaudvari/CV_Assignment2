```python
from pathlib import Path

import matplotlib.pyplot as plt

# assignment root when chest_xray is not inside this commit folder
nb_dir = Path.cwd().resolve()
root = nb_dir if (nb_dir / "chest_xray").is_dir() else nb_dir.parent.parent
train_dir = root / "chest_xray" / "train"
test_dir = root / "chest_xray" / "test"

from pneumonia_classification import count_files_per_class

train_c = count_files_per_class(train_dir)
test_c = count_files_per_class(test_dir)
print("Train:", train_c)
print("Test:", test_c)

labels = list(train_c.keys())
vals = [train_c[k] for k in labels]
plt.figure(figsize=(7, 4))
plt.bar(labels, vals, color=["#4c72b0", "#55a868", "#c44e52"])
plt.title("Training images per class (imbalanced)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
